```python
import numpy as np

#driverStandingsWin = driverStandings.groupby('driverId', as_index=False)['wins'].sum()#.max()#.count()
#driverStandingsWin

Winners = results.loc[results['positionOrder'] == 1]
WinnersModified = Winners.copy()
WinnersModified = WinnersModified.drop(columns = ['laps','time', 'milliseconds', 'fastestLap', 'fastestLapTime', 'fastestLapSpeed','positionText','positionOrder','rank','points'])
WinnersModified.head()
```

```python
DriversWinAmount = Winners.groupby('driverId', as_index=False)['position'].sum()
DriversWinAmount.sort_values(['position'], axis=0, ascending=False, inplace = True)
DriversWinAmount.rename(columns = {'position' : 'TotalWins'}, inplace = True)
DriversWinAmount.head()
```

```python
#Winners = DriversWinAmount.merge(Winners, how='inner')
WinnersModified = WinnersModified.merge(DriversWinAmount, how='inner')
WinnersModified.head()
```

```python
drivers = drivers[['driverId', 'nationality', 'dob','forename','surname']]
drivers.head()
```

```python
countries = filecontent[4]
countries.drop(['ISO','Country'], axis=1, inplace = True)
countries.rename(columns = {'Nationality' : 'nationality'}, inplace = True)
countries.head()
```

```python
WinnersModified = drivers.merge(WinnersModified, how='inner')
WinnersModified = WinnersModified.merge(countries, how ='inner')
WinnersModified = WinnersModified.merge(constructors, how ='inner')
WinnersModified = racesDate.merge(WinnersModified, how ='inner')
WinnersModified.head()
```

```python
WinnersModified = WinnersModified.drop(columns = ['raceId','resultId','constructorId','number','statusId'])
```

```python
WinnersModified.head()
```

```python
WinnersModified['date'] = pd.to_datetime(WinnersModified['date'])
```

```python
#Drivers with more amount of accidents quit the f1 earlier than drivers with less amount of accidents
dfAccidents = results[['resultId', 'raceId', 'driverId', 'constructorId','laps', 'statusId']]

#exceptionlaps = [1,range(11,19)]
exceptions1 = [x for x in range(11,20)]
exceptions2 = [1,45,50,128,53,55,58,88,122,123,124,125,127,133,134]
exceptions3 = [x for x in range(111,121)]
exceptions = exceptions1 + exceptions2 + exceptions3


RaceIsNotFinished = dfAccidents.loc[~dfAccidents['statusId'].isin(exceptions)]
RaceIsFinished = dfAccidents.loc[dfAccidents['statusId'].isin(exceptions)]

RaceIsFinishedLaps = RaceIsFinished.groupby('driverId').count()['raceId'].copy()
RaceIsFinishedLaps.sort_values(ascending=False, inplace = True)

RaceIsNotFinishedLaps = RaceIsNotFinished.groupby('driverId').count()['raceId'].copy()
RaceIsNotFinishedLaps.sort_values(ascending=False, inplace = True)

RaceTotal = dfAccidents.groupby('driverId').count()['raceId'].copy()
RaceTotal.sort_values(ascending=False, inplace = True)

result = pd.concat([RaceIsFinishedLaps.rename('FinishedRaces'), RaceIsNotFinishedLaps.rename('NotFinishedRaces'), RaceTotal.rename('TotalRaces')], axis=1)
result.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
result.reset_index(inplace = True)
result.head()
```

```python
Winners = WinnersModified.merge(result, how='inner')
Winners.head()
```
